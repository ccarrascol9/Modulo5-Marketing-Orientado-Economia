```python
df = df[["date", "time", "username", "replies_count","retweets_count","likes_count","tweet"]]
```

```python
df
```

```python
df['promedio']=(df['replies_count']+df['retweets_count']+df['likes_count'])/3
df.head(5)
```

```python
condiciones = [ (df['promedio']<10),
               (df['promedio']>10) & (df['promedio']<100),
               (df['promedio']>100)
    
]
opciones = ['negativo','neutro','positivo']
df['clasificac']=np.select(condiciones,opciones)
df.head(20)
```

```python
negativas=["Europa", 'Internacional', 'Sandro']
positivas=['boric', 'izkia', 'Rechazo']
```

```python
df['negativas']=df.tweet.str.count('|'.join(negativas))
```

```python
df['positivas']=df.tweet.str.count('|'.join(positivas))
```
